```python
subtopics = data['Sub Topics'].apply(lambda x: pd.Series(str(x).split(",")))
```

```python
subtopics
```

```python
subtopics.fillna('nan',inplace=True)
subtopics.dropindex
```

```python
subtopics
```

```python
tops4 = []
subs4 = []

x1 = [j for j in subtopics.values]
for k in range(0,len(x1)):
    for l in range(len(x1[k])):
        #print(l)
        if x1[k][l] != 'nan':
            tops4.append(data['Topics'][k])
            #print(x1[k][l])
            subs4.append(x1[k][l])
        #print("-----")
    #print("******")
```

```python
df = pd.DataFrame(list(zip(tops4, subs4)), columns =['Topics', 'Sub Topics']) 
```
